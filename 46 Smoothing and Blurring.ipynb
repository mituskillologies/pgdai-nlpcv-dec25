#### Averaging Kernel

```python
img = cv2.imread('images/dog.jpg')
```

```python
kernel = np.zeros([3,3]) + 1/9
```

```python
kernel
```

```python
blurred_image = cv2.filter2D(img, -1, kernel)
```

```python
cv2.imshow("blurred image", blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.zeros([9,9]) + 1/81
blurred_image = cv2.filter2D(img, -1, kernel)
blurred_image = np.hstack([img, blurred_image])
```

```python
cv2.imshow("blurred image", blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Gaussian Kernel

```python
kernel_size = 7
kernel = cv2.getGaussianKernel(kernel_size, 1.23)
```

```python
kernel
```

```python
gaussian_blurred_image = cv2.filter2D(img, -1, kernel)
gaussian_blurred_image = np.hstack([img, gaussian_blurred_image])
```

```python
cv2.imshow("gaussian blurred image", gaussian_blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Median Blur

```python
img_gray = cv2.imread('images/dog.jpg', 0)
```

```python
median_img = cv2.medianBlur(img_gray, 15) # 15 -> Kernel Size
```

```python
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.imshow(img_gray, cmap = 'gray')
plt.subplot(122)
plt.imshow(median_img, cmap = 'gray')
```

#### Blur the patch

```python
gray_copy = img_gray.copy()
patch = img_gray[100:250, 100:250] 
blurred_patch = cv2.medianBlur(patch, 15)
gray_copy[100:250, 100:250] = blurred_patch
```

```python
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.imshow(img_gray, cmap = 'gray')
plt.subplot(122)
plt.imshow(gray_copy, cmap = 'gray')
```
