#### Extract specific POS from the word

```python
verbs = set()
nouns = set()
adjts = set()
# Extracts the synonyms -> sound
for word in wordnet.synsets('sound', pos='n'):
    nouns.add(word.lemmas()[0].name())
for word in wordnet.synsets('sound', pos='v'):
    verbs.add(word.lemmas()[0].name())
for word in wordnet.synsets('sound', pos='a'):
    adjts.add(word.lemmas()[0].name())
```

#### Retrieve the antonyms

```python
antonyms = set()

for word in wordnet.synsets('possible'):
    for lemma in word.lemmas():
        if lemma.antonyms(): # Check if antonyms is present
            antonyms.add(lemma.antonyms()[0].name())
```

##### Extract the hypernyms

```python
word = 'intelligence'
synsets = wordnet.synsets(word)

for x in synsets:
    hypernyms = x.hypernyms()
    for h in hypernyms:
        print(word,'hypernym of', h.name())
```

##### Extract meronyms

```python
word = 'car'
synsets = wordnet.synsets(word)

for x in synsets:
    meronyms = x.part_meronyms()
    for m in meronyms:
        print(word,'meronym of', m.name())
```

```python
word = 'computer'
synsets = wordnet.synsets(word)

for x in synsets:
    meronyms = x.part_meronyms()
    for m in meronyms:
        print(word,'meronym of', m.name())
```

#### Extract holonyms
