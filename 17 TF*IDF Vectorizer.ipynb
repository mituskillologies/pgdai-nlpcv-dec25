```python
sent1 = 'It is good practice for us.'
sent2 = 'It was also good to know about it.'
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
new_text = tfidf.fit_transform([sent1, sent2])
```

```python
new_text.toarray()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(new_text.toarray(), columns=tfidf.get_feature_names_out())
```

```python
df
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
new_text = tfidf.fit_transform([sent1, sent2])
```

```python
df = pd.DataFrame(new_text.toarray(), columns=tfidf.get_feature_names_out())
```

```python
df
```

```python
f = open('sample_sentences.txt','r')
new_text = tfidf.fit_transform(f.readlines())
df = pd.DataFrame(new_text.toarray(), columns=tfidf.get_feature_names_out())
df
```
