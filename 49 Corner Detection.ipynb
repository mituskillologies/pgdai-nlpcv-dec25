```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('images/circle.png', 0)
```

```python
help(cv2.cornerHarris)
```

```python
corners = cv2.cornerHarris(img, blockSize= 2, ksize = 3, k = 0.04)
```

```python
corners.min()
```

```python
corners.shape
```

```python
plt.imshow(corners)
```

```python
corners.min(), corners.max(), np.median(corners)
```

```python
img_new = img.copy()
img_new[corners > 0.01 * corners.max()] = [255]
```

```python
plt.imshow(img_new, cmap = 'gray');
```

```python
img_new = corners < corners.max() / 2
```

```python
plt.imshow(img_new, cmap = 'gray')
```
