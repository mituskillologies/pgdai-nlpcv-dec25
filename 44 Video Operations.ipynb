```python
import cv2
import matplotlib.pyplot as plt
```

```python
# load the video 
vid = cv2.VideoCapture('images/nature.mp4')
```

```python
ret, frame = vid.read()
```

```python
cv2.imshow("First Frame", frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Extract video metadata 
width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
total_frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
fps = int(vid.get(cv2.CAP_PROP_FPS))
```

```python
print("Width:", width)
print("Height:", height)
print("Frame Count:", total_frames)
print("Frame Rate:", fps)
```

```python
duration = total_frames / fps
```

```python
duration
```

```python
# Display the video 
vid = cv2.VideoCapture('images/nature.mp4')

while True:
    ret, frame = vid.read()
    if ret == False:
        break
    cv2.imshow("My Video", frame)
    if cv2.waitKey(5) == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
cv2.imwrite("images/myframe.jpg", frame)
```

```python
# Display the video 
vid = cv2.VideoCapture('images/nature.mp4')
i = 1
while True:
    ret, frame = vid.read()
    if ret == False:
        break
    cv2.imwrite("images/frames/frame%d.jpg" %i, frame)
    i += 1
```

```python
time = "00:00:45.00"
```

```python
hrs, mins, secs = time.split(":")
```

```python
hrs, mins, secs
```

```python
trigger_frame = int(hrs) * 3600 * fps + int(mins) * 60 * fps + float(secs) * fps
```

```python
trigger_frame
```

```python
vid.set(1, trigger_frame)
```

```python
ret, frame = vid.read()
cv2.imshow("Trigger Frame", frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
