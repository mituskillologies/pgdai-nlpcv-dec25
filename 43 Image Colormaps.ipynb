```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/dog.jpg', 0)
```

```python
orig_map = plt.colormaps['Greys'].reversed()
```

#### Sequential

```python
plt.figure(figsize=(16,16))
plt.subplot(441)
plt.imshow(img, cmap = plt.colormaps['Greys'].reversed()); plt.title('Greys')
plt.subplot(442)
plt.imshow(img, cmap = plt.colormaps['Blues'].reversed()); plt.title('Blues')
plt.subplot(443)
plt.imshow(img, cmap = plt.colormaps['Reds'].reversed()); plt.title('Reds')
plt.subplot(444)
plt.imshow(img, cmap = plt.colormaps['Greens'].reversed()); plt.title('Greens')
plt.subplot(445)
plt.imshow(img, cmap = plt.colormaps['Oranges'].reversed()); plt.title('Oranges')
plt.subplot(446)
plt.imshow(img, cmap = plt.colormaps['Purples'].reversed()); plt.title('Purples')
plt.subplot(447)
plt.imshow(img, cmap = plt.colormaps['PuRd'].reversed()); plt.title('PuRd')
plt.subplot(448)
plt.imshow(img, cmap = plt.colormaps['YlGn'].reversed()); plt.title('YlGn')
```

#### Diverging

```python
plt.figure(figsize=(16,16))
plt.subplot(441)
plt.imshow(img, cmap = 'PiYG'); plt.title('PiYG')
plt.subplot(442)
plt.imshow(img, cmap = 'PRGn'); plt.title('PRGn')
plt.subplot(443)
plt.imshow(img, cmap = 'BrBG'); plt.title('BrBG')
plt.subplot(444)
plt.imshow(img, cmap = 'PuOr'); plt.title('PuOr')
plt.subplot(445)
plt.imshow(img, cmap = 'RdGy'); plt.title('RdGy')
plt.subplot(446)
plt.imshow(img, cmap = 'RdBu'); plt.title('RdBu')
plt.subplot(447)
plt.imshow(img, cmap = 'Spectral'); plt.title('Spectral')
plt.subplot(448)
plt.imshow(img, cmap = 'coolwarm'); plt.title('coolwarm')
```

#### Cyclic

```python
plt.figure(figsize=(16,16))
plt.subplot(441)
plt.imshow(img, cmap = 'hsv'); plt.title('hsv')
plt.subplot(442)
plt.imshow(img, cmap = 'twilight'); plt.title('twilight')
plt.subplot(443)
plt.imshow(img, cmap = 'twilight_shifted'); plt.title('twilight_shifted')
plt.subplot(444)
plt.imshow(img, cmap = 'viridis'); plt.title('viridis')
```
