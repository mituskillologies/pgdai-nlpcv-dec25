```python
words1 = 'consultant','consult','consultants','consulting','consultative'
words2 = 'management','managing','managed','manage'
words3 = 'big','bigger','biggest'
words4 = 'go','gone','went','going'
words5 = 'association','associating','associated','associations'
```

```python
# Import the stemmer
from nltk.stem import PorterStemmer
```

```python
# create the object
ps = PorterStemmer()
```

```python
ps.stem(words2[0])
```

```python
for word in words1:
    print(ps.stem(word))
```

```python
for word in words2:
    print(ps.stem(word))
```

```python
for word in words3:
    print(ps.stem(word))
```

```python
for word in words4:
    print(ps.stem(word))
```

```python
from nltk.stem import LancasterStemmer
```

```python
ls = LancasterStemmer()
```

```python
for word in words1:
    print(ls.stem(word))
```

```python
for word in words2:
    print(ls.stem(word))
```

```python
for word in words3:
    print(ls.stem(word))
```

```python
for word in words4:
    print(ls.stem(word))
```

```python
for word in words5:
    print(ls.stem(word))
```
