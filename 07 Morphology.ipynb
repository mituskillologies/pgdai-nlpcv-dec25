```python
from polyglot.text import Word
```

```python
words = ['cats','computing','association','programming','leadership',
        'miscommunication','identifiable','psychologically','engineering',
        'tabular','realistic','colorfulness']
```

```python
for w in words:
    w = Word(w, language='en')
    print(w,'\t\r', w.morphemes)
```

```python
mwords = ['राज्यशास्त्र','जलविद्युत','महाद्वीप','मानसशास्त्रीय']
for word in mwords:
    w = Word(word, language='mr')
    print(word,'\t\t', w.morphemes)
```
