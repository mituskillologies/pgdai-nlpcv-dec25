```python
text = soup.get_text()
```

```python
print(text)
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(text)
```

```python
tokens = " ".join(tokens)
```

```python
tokens
```

```python
from nltk import FreqDist
```

```python
freq = FreqDist(word_tokenize(tokens))
```

```python
freq.plot(20)
```

```python
tokens1 = word_tokenize(tokens)
```

```python
# remove punctuations
```

```python
tokens2 = [t for t in tokens1 if t.isalpha() or t.isdigit()]
```

```python
# remove stopwords
from nltk.corpus import stopwords
swords = stopwords.words('english')
```

```python
tokens3 = [t.lower() for t in tokens2 if t.lower() not in swords]
```

```python
freq = FreqDist(tokens3)
freq.plot(20)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
tokens4 = [ps.stem(t) for t in tokens3]
```

```python
freq = FreqDist(tokens4)
freq.plot(20)
```

```python
url = 'https://en.wikipedia.org/wiki/Rajgad'
headers = {
    'User-Agent':'Custom-Script/1.9 (tushar@gmail.com)',
    'Accept-Language': 'en-US,en;q=0.5'
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text()
tokens = word_tokenize(text)
tokens = " ".join(tokens)
tokens1 = word_tokenize(tokens)
tokens2 = [t for t in tokens1 if t.isalpha() or t.isdigit()]
tokens3 = [ps.stem(t.lower()) for t in tokens2 if t.lower() not in swords]
freq = FreqDist(tokens3)
freq.plot(20)
```
