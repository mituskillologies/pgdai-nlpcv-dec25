```python
from gtts import gTTS
from playsound import playsound
```

```python
# create the object of gtts
tts = gTTS(text='Welcome to the world of Deep Learning using Python.',
           lang='en')
```

```python
# save the audio
tts.save('audio.mp3')
playsound('audio.mp3')
```
