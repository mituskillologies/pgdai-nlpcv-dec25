```python
char_to_int = dict((v, i) for i, v in enumerate(alphabets))
```

#### Prepare the dataset

```python
seq_length = 1
x = []
y = []

for i in range(len(alphabets) - seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1] 
    print(seq_in,'-->',seq_out)   
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
```

```python
import numpy as np
```

```python
x = np.reshape(x, (25, 1, 1))
```

```python
x_norm = x / 25
```

```python
from keras.utils import to_categorical
```

```python
a = [1,3,2,3,2,1,1,2,3]
```

```python
y_new = to_categorical(y)
```

```python
y_new.shape
```

```python
x_norm.shape
```

#### Build the model

```python
from keras.layers import SimpleRNN, Dense, Input
from keras.models import Sequential
```

```python
model = Sequential()

model.add(Input((1,1)))
model.add(Dense(25, activation='relu'))
model.add(SimpleRNN(32))
model.add(Dense(26, activation='softmax'))
```

```python
model.summary()
```

#### Compile the model

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adam',metrics= ['accuracy'])
```

#### Train the model

```python
model.fit(x_norm, y_new, epochs=500, batch_size=1)
```

```python
model.evaluate(x_norm, y_new, batch_size=1)
```

```python
new = 'M'
new = char_to_int[new]
new = np.reshape(new,(1,1,1))
new = new / 25
pred = model.predict(new, verbose=False)
pred = np.argmax(pred)
int_to_char[pred]
```

```python
seq_length = 3
x = []
y = []

for i in range(len(alphabets) - seq_length):
    seq_in = alphabets[i:i+seq_length]
    seq_out = alphabets[i+seq_length] 
    print(seq_in,'-->',seq_out)   
    x.append([char_to_int[char] for char in seq_in])
    y += [char_to_int[seq_out]]
```

```python
x_new = np.reshape(x, (23, seq_length, 1))
```

```python
y_new = to_categorical(y)
y_new.shape
```

```python
x_new = x_new / 25
```

```python
x_new;
```

```python
model = Sequential()

model.add(Input((3,1)))
model.add(Dense(25, activation='relu'))
model.add(SimpleRNN(32))
model.add(Dense(26, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',
             metrics=['accuracy'])
```

```python
model.fit(x_new, y_new, epochs=1000, batch_size=1)
```

```python
new = 'JKL'
val = np.array([[char_to_int[x]] for x in new])
val = val.reshape(1,3,1)
val = val / 25
result = model.predict(val, verbose=False)
result = np.argmax(result)
print(new,'-->',int_to_char[result])
```
