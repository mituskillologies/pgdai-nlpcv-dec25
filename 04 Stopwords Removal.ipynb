```python
sent = 'Hello, friends! How are you? Welcome to "Python" Programming.'
```

```python
sent
```

```python
from nltk.tokenize import word_tokenize
import string
```

```python
tokens = word_tokenize(sent)
```

```python
clean = [word for word in tokens if word.isalpha() or word.isdigit()]
```

```python
" ".join(clean)
```

```python
new_sent=''
for char in sent:
   if char.isalpha() or char.isdigit() or char.isspace():
       new_sent += char
```

```python
new_sent
```

##### Stopwords Removal

```python
from nltk.corpus import stopwords
```

```python
swords = stopwords.words('english')
```

```python
swords
```

```python
clean_text = [word for word in word_tokenize(new_sent) if word.lower() not in swords]
```

```python
" ".join(clean_text)
```
