```python
sent = 'They told that their ages are 26 32 31 and 29 respectively.'
```

```python
# Write a program to find the average of ages mentioned in the sentence.
# Value and count of ages can be altered
```

```python
words = sent.split()
```

```python
num = []
for word in words:
    if word.isdigit():
        num.append(int(word))
```

```python
num
```

```python
sum(num) / len(num)
```

```python
num = [int(word) for word in sent.split() if word.isdigit()]
```

```python
sum(num) / len(num)
```

##### Cuctom Tokenizer

```python
import re
```

```python
def custom_tokenizer(text):
    return re.split(r"[.,;?!\s]+", text)

text = "This is some text with punctuation > Let's tokenize it. Is it ok?"
```

```python
custom_tokenizer(text)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("The cat purred softly.")

for token in doc:
    print(f"Word: {token.text}")
    print(f"  Lemma: {token.lemma_}")
    print(f"  PoS: {token.pos_}")
    print(f"  Morphology: {token.morph}")
    print("-" * 15)
```
