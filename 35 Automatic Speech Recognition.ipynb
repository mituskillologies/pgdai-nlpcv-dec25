### TTS

```python
import pyttsx3
```

```python
engine = pyttsx3.init()
```

```python
# Set the properties
engine.setProperty('rate', 100)
engine.setProperty('volume', 0.9)
```

```python
text = 'Welcome to world of Deep Learning using Python Programming.'
```

```python
engine.say(text)
engine.runAndWait()
```

```python
r = sr.Recognizer()
```

```python
with sr.Microphone() as source: 
    r.adjust_for_ambient_noise(source)
    print('Speak:')
    audio = r.listen(source) 
    print('You said:', r.recognize_google(audio))
```
