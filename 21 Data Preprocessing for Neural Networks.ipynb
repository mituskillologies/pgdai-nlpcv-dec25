```python
lines = [
    'It was a nice rainy day.',
    'The things were so beautiful in his point.',
    'When your focus is clear, you won.',
    'Many many happy returns of the day.'
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(lines)
```

```python
tokenizer.word_docs
```

```python
tokenizer.index_word
```

```python
tokenizer.index_docs
```

```python
tokenizer.word_index
```

```python
mat = tokenizer.texts_to_matrix(lines)
```

```python
mat
```

```python
tokenizer.word_index.keys()
```

```python
mat.shape
```

```python
seq = tokenizer.texts_to_sequences(lines)
```

```python
seq
```

```python
padded = pad_sequences(seq, maxlen=10, padding='post')
```

```python
padded
```

```python
padded = pad_sequences(seq, maxlen=10, padding='pre')
```

```python
padded
```

```python
m = 0
for sent in seq:
    if len(sent) > m:
        m = len(sent)
```

```python
m
```

```python
padded = pad_sequences(seq, maxlen=m, padding='post')
```
