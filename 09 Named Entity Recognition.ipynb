```python
nlp = spacy.load('en_core_web_sm')
```

```python
nlp
```

```python
sent = nlp('''Mark Zukerberg will meet Aditya Joshi
on Monday 25 November 2025, 10am for $3 Trillion deal
at Pune for second time.''')
```

```python
print(sent)
```

```python
sent.ents
```

```python
for ent in sent.ents:
    print(ent.text,' ---> ', ent.label_)
```
