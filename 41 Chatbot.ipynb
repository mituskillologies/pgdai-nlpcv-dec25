```python
import nltk
import random
```

```python
# define the chatbot responses and intents 
responses = {
    'greeting': ['Hello!','Hi there!','Hey!','Howdy!'],
    'goodbye': ['Good bye!','See you later!','Take Care','Bye!'], 
    'thanks': ['You\'re welcome.','Glad to help','Anytime','Happy to assist'], 
    'default': ['Sorry I didnt understand.','Can you tell again?', 
               'I am not sure.']
}
```

```python
# simple set of questions and responses or patterns 
patterns = {
    'hi':'greeting',
    'hello':'greeting',
    'hey':'greeting',
    'bye':'goodbye',
    'thanks':'thanks', 
    'thank you':'thanks'
}
```

```python
# Match the user 
def respond(user_input):
    for word in nltk.tokenize.word_tokenize(user_input.lower()): 
        if word in patterns: 
            intent = patterns[word] 
            return random.choice(responses[intent])
    return random.choice(responses['default'])
```

```python
respond('hi')
```

```python
def chat(): 
    print('Chatbot: How can I help you today? (Type "exit" to quit)')
    while True:
        user_input = input('You:')
        if user_input == 'exit': 
            print('Chatbot: Good Bye!')
            break 
        response = respond(user_input)
        print('Chatbot: ', response)
```

```python
chat()
```
